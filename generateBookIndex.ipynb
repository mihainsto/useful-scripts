```python
import PyPDF2
from tqdm import tqdm
```

```python
file = open('soYale.pdf', 'rb')
```

```python
fileReader = PyPDF2.PdfFileReader(file)
```

```python
print(fileReader.numPages)
pageObj = fileReader.getPage(5)
pageObj.extractText()
```

```python
def get_words_from_page(pageText):
    pageText = pageText.split('\n')
    allWords = []
    for line in pageText:
        for word in line.split(' '):
            if word != '':
                allWords.append(word)
    
    return allWords

def clean_word(sample):
    to_replace = """[,\.!?'"1234567890]-():;}{}=/*"""
    for char in to_replace:
        sample = sample.replace(char,'')
    return sample
```

```python
wordsDict = {}
wordPages = {}
```

```python
for i in tqdm(range(0, fileReader.numPages)):
    pageObj = fileReader.getPage(i)
    words = get_words_from_page(pageObj.extractText())
    for word in words:
        cleaned_word = clean_word(word)
        if len(cleaned_word) <= 2:
            continue
        try:
            wordsDict[cleaned_word] += 1
            wordPages[cleaned_word].append(i)
        except:
            wordsDict[cleaned_word] = 1
            wordPages[cleaned_word] = [i]
```

```python
wordsList = list(wordsDict.items())
```

```python
wordsList.sort(key = lambda x:x[1])
```

```python
wordsList = wordsList[:9000]
```

```python
finalWords = []
```

```python
for word in wordsList:
        finalWords.append((word[0], wordPages[word[0]]))

```

```python
newFinalWords = []
for word in finalWords:
    if len(word[1]) < 20:
        newArr = []
        for el in word[1]:
            newArr.append((el//6)+1)
        newWord = (word[0], newArr)
        newFinalWords.append(newWord)
        
```

```python
with open('indexNew.txt','a', encoding="utf-8") as f:
    for word in newFinalWords:
        toWriteWord = str(word)
        toWriteWord=toWriteWord.replace("'","")
        toWriteWord=toWriteWord.replace("(","")
        toWriteWord =toWriteWord.replace(")","")
        f.write(toWriteWord)
        f.write('\n')
```
